Pratyush has a list of items (Harry Potter, Gone Girl, etc.) he wants to complete this Summer and has stored the genres of each item (Fantasy Fiction, Mystery, etc.), in several dictionaries based on the type of activity (Books, Movies or Video games).  These dictionaries can be defined using the code snippet given below.  

books = {'Harry Potter': 'Fantasy Fiction', 'Gone Girl': 'Mystery'}
movies = dict([('Seabiscuit','Drama/Sport'), ('House of Wax', 'Horror/Thriller')])
video_games = dict(StardewValley='Adventure', Fortnite='Survival')

			
			Format (just to clear any confusion): Activity = {‘Item’: ‘Genre’}

Write code that will do the following:
1.	Print out all of Pratyush’s Summer items.
2.	Now print all of their genres.
3.	Now print them as ‘pairs’ (item, genre).
4.	Pratyush no longer wants to read the book ‘Gone Girl’. Remove it from the books list.
5.	Pratyush needs to send the whole list to their friend.  Make a single dictionary containing all the items. (Use this dictionary for further questions)
6.	Pratyush decided to watch the comedy movie ‘Cruella’ based on their brother’s suggestion. Add it to the dictionary!
7.	Pratyush wants to know what genre the movie ‘House of Wax’ falls under. Get the genre.
8.	Summer’s over and Pratyush is finally done with everything. Back to school now. Remove everything from the Dictionary.


```python
books = {'Harry Potter': 'Fantasy Fiction', 'Gone Girl': 'Mystery'}
movies = dict([('Seabiscuit','Drama/Sport'), ('House of Wax', 'Horror/Thriller')])
video_games = dict(StardewValley='Adventure', Fortnite='Survival')

print("\n","1.Pratyush's Summer Items are:",'\n')
for i in books.keys():
    print(i)
for i in movies.keys():
    print(i)
for i in video_games.keys():
    print(i)
    
print("\n","2. All of their respective genres are:","\n")

for i in books.values():
    print(i)
for i in movies.values():
    print(i)
for i in video_games.values():
    print(i)

print("\n","3.We can see their pairs as:","\n")

for i,j in books.items():
    print(i,":",j)
for i,j in movies.items():
    print(i,":",j)
for i,j in video_games.items():
    print(i,":",j)

print("\n","4 .Updated Book List:")
del books["Gone Girl"]
for i in books.keys():
    print(i)
    
print("\n","5.Single list of all the items are as follows:")
collection=dict(**books,**movies,**video_games)
print(collection)

print("\n","6.Updated list with brothers suggestion")
collection['Cruella']='Comedy'
print(collection)

print("\n","7. The genre of House of wax is ",collection['House of Wax'],"\n")

print("\n","8. Summer is over,the items remaining are",collection.clear())
```
